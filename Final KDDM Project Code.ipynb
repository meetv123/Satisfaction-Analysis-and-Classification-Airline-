```python
# Replace white spaces from column names
train.columns = train.columns.str.replace(" ", "_")
test.columns = test.columns.str.replace(" ", "_")
train.columns
```

```python
train.Gender[train.Gender == 'Male'] = 1
train.Gender[train.Gender == 'Female'] = 2
train.Customer_Type[train.Customer_Type =="Loyal Customer"]=1
train.Customer_Type[train.Customer_Type =="disloyal Customer"]=2
train.Type_of_Travel[train.Type_of_Travel =="Personal Travel"]=1
train.Type_of_Travel[train.Type_of_Travel =="Business travel"]=2
train.Class[train.Class =="Business"]=1
train.Class[train.Class =="Eco"]=2
train.Class[train.Class =="Eco Plus"]=3

print(train)
```

```python
# Converting string values in columns to numeric values to use in classification models
test.Gender[test.Gender == 'Male'] = 1
test.Gender[test.Gender == 'Female'] = 2
test.Customer_Type[test.Customer_Type =="Loyal Customer"]=1
test.Customer_Type[test.Customer_Type =="disloyal Customer"]=2
test.Type_of_Travel[test.Type_of_Travel =="Personal Travel"]=1
test.Type_of_Travel[test.Type_of_Travel =="Business travel"]=2
test.Class[test.Class =="Business"]=1
test.Class[test.Class =="Eco"]=2
test.Class[test.Class =="Eco Plus"]=3

test
```

```python
def normalizeFeatures(X):
    mu=np.mean(X)
    sigma=np.std(X)
    X_norm=(X - mu)/sigma
    return X_norm
```

```python
X_train=train.iloc[:,:-1] #independent values
Y_train=train.iloc[:,-1] #dependent value
X_train
```

```python
X_test=test.iloc[:,:-1] #independent values
Y_test=test.iloc[:,-1] #dependent value
X_test
```

```python
normalized_train=normalizeFeatures(X_train)
normalized_train
# normalized_test=normalizeFeatures(X_test)
# normalized_test
```

```python
Y_train = train['satisfaction'].map({'neutral or dissatisfied': 0, 'satisfied': 1})
Y_test = test['satisfaction'].map({'neutral or dissatisfied': 0, 'satisfied': 1})
Y_train
```

```python
# Import random forest classifier
from sklearn.ensemble import RandomForestClassifier

# Train and optimize random forest classifier
rfc = RandomForestClassifier(n_estimators=100)

# Fit the model
rfc.fit(normalized_train, Y_train)

# Print the score
print(f"Testing score : {rfc.score(normalized_test, Y_test)*100}")
```

```python
y_pred=rfc.predict(X_test)
print(y_pred)
```

```python
Result3=pd.DataFrame({"Actual": Y_test, " Predicted": y_pred})
Result3.sort_index()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(Y_test,y_pred))
print(confusion_matrix(Y_test,y_pred))
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(rfc, normalized_test, Y_test)  
plt.show()
```

NAIVE BAYES 

```python
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(normalized_train, Y_train)

y_pred_nb = nb_classifier.predict(normalized_test)

# Lets measure the accuracy of the model using the confusion matrix.

print(f"Testing score : {nb_classifier.score(normalized_test, Y_test)*100}")
```

```python
cm = confusion_matrix(Y_test, y_pred_nb)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(Y_test,y_pred_nb))
print(confusion_matrix(Y_test,y_pred_nb))
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(nb_classifier, normalized_test, Y_test)  
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(criterion = 'entropy')
dt_classifier.fit(normalized_train, Y_train)

y_pred_dt = dt_classifier.predict(normalized_test)



print(f"Testing score : {dt_classifier.score(normalized_test, Y_test)*100}")
```
